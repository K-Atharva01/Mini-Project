```python
import numpy as np
import cv2
import os
import random
import matplotlib.pyplot as plt
from tensorflow import keras
```

```python
model = keras.models.load_model(r'my_model.h5')
```

```python
DIR = r'D:\Mini Project SE\Test'
img_size = 100

test_data = []


for img in os.listdir(DIR): 
    img_path = os.path.join(DIR,img) 
    img_arr = cv2.imread(img_path) 
    img_arr = cv2.resize(img_arr,(100,100))
    test_data.append(img_arr)
```

```python
test = np.array(test_data, dtype = np.float32)
test = test/255
```

```python
y = model.predict(test)
```

```python
y
```

```python
def numbers_to_strings(argument):
    switcher = {
        0: "cat",
        1: "chicken",
        2: "cow",
        3: "dog",
        4: "elephant",
        5: "horse",
        6: "squirrel"
    }
    return switcher.get(argument, "nothing")
    
for img in y:
    a = np.argmax(img)
    print(numbers_to_strings(a))
    
    

        
```
